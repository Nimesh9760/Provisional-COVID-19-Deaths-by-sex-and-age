```python
#Average death ratio occured due to all the diseases in all the states
averages=covid_deaths.groupby('State')['Pneumonia_Influenza_o_ COVID-19_Deaths'].mean()
print(averages)
```

```python
print(type(averages))
```

```python
averages = covid_deaths.groupby('State')['Pneumonia_Influenza_o_ COVID-19_Deaths'].mean().to_frame(name = 'mean').reset_index()
print(averages)
```

```python
print(type(averages))
```

```python
averages.sort_values("mean", ascending=False,
                     inplace = True, na_position ='first')
print(averages)
```

```python
from pandas.core.groupby. grouper import DataFrame
import plotly.express as px
fig1 = px.box(covid_deaths, x="State", y="Pneumonia_Influenza_o_ COVID-19_Deaths")
fig1.show()
```

```python
#Maximum total death ratio over all the states
from numpy.ma import maximum
maximum = covid_deaths.groupby('State')['Total_Deaths'].max()
print(maximum)
```

```python
print(type(maximum))
```

```python
from numpy.ma import maximum
maximum = covid_deaths.groupby('State')['Total_Deaths'].max().to_frame(name = 'max').reset_index()
print(maximum)
```

```python
print(type(maximum))
```

```python
maximum.sort_values("max", ascending=True,
                 inplace = True, na_position ='first')
print(maximum)
```

```python
x=maximum.iloc[:,0].values
print(x)
y=maximum.iloc[:,1].values
print(y)
plt.figure(figsize=(110, 50))
display(plt.bar(x, y))
plt.bar(x,y)
plt.show()
```
